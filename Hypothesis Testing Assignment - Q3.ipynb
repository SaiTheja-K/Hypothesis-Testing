## Hypothesis Testing Exercise :

## 3) Sales of products in four different regions is tabulated for males and females. Find if male-female buyer rations are similar across regions.

```python
import pandas as pd
import numpy as np
from scipy import stats
from scipy.stats import chi2_contingency
```

```python
buyer_data =pd.read_csv("BuyerRatio.csv")
```

```python
buyer_data
```

```python
buyer_data.drop(['Observed Values'],axis=1,inplace = True)
```

```python
buyer_data
```

```python
# Make dimensional array
obs_val=np.array([[50,142,131,70],[435,1523,1356,750]])
obs_val
```

```python
chi2,p_val,dof,expected = chi2_contingency(obs_val)
```

```python
print('Chi-squared value : ',round(chi2,4))
print('P-val             : ',round(p_val,4))
print('Degree of Freedom : ',dof)
print('Expected Table    :\n',expected)
```

# P_Value = 0.66 (66%).

# HYPOTHESIS FORMULATION:

# H0: All proportions are equal.

# Ha: Not all proportions are equal.


# Level of significance : 5%. (0.05)

# Since the p_value is 0.66 (66%) > than the level of significance 0.05(5%) (p_value > LOS)

# so, we do not reject the Null Hypothesis (H0).


# Observation - All proportions are equal .
